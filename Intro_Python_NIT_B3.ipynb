```python
#WAP code to check whether a year is leap
#year or not using function

#function defination
def fun(year):
    if((year %4==0 and year %100!=0) or (year %400==0)):
        return True
    else:
        return False

#Function call
year=int(input("Enter the year"))
ans=fun(year)
print(ans)
```

```python
#Function with loops
#WAP code to print the sum of first N natutal number using function
def fun(n):
    sum=0
    for i in range(1,n+1,1):
        sum+=i
    return sum

ans=fun(10)
print(ans)
```

```python
#Properties for the List
#List:-Ordered,Changable,Duplicate will be allowed
l=["Virat","Rohit","Suresh","Jadeja"]
print(l)#Follow the order
s={"Virat","Rohit","Suresh","Jadeja"}
print(s)#Order is not Following
```

```python
#Changable Properties
l=["Virat","Rohit","Suresh","Jadeja"]
l[0]="Sachin"
print(l)
```

```python
#Allows Duplicate
l=["Virat","Rohit","Suresh","Jadeja","Virat"]
print(l)
```

```python
#Insert Operation in List
'''
append(),It is use to insert the elements from tail end
insert(),For the position wise inserting the insert will be use
extend(),Use to merge the two list

'''
l=["Virat","Rohit","Suresh"]
l.append("Yuvraj")
print(l)
```

```python
l=["Virat","Rohit","Suresh"]
l.insert(1,"Yuvraj")
print(l)
```

```python
l1=["Virat","Rohit","Suresh"]
l2=["Yuvraj","Jadeja"]
l1.extend(l2)
print(l1)
```

```python
#Delete operation in list
'''
remove(),For element wise delete
pop()
del
clear


'''
l=["Virat","Rohit","Suresh","Jadeja"]
l.remove("Rohit")
print(l)
```

```python
#For the index wise remove
l=["Virat","Rohit","Suresh","Jadeja"]
l.pop(2)
print(l)
```

```python
#Clear
l=["Virat","Rohit","Suresh","Jadeja"]
l.clear()
print(l)
```

```python
#del
l=["Virat","Rohit","Suresh","Jadeja"]
del l
print(l)
```

```python
#List Comprehension
l=[1,2,3,4,5,6]
ans=[i for i in l if i%2==0]
print(ans)
```

```python
#Sorting in list
l=[10,21,0,-1,6]
l.sort(reverse=True)
print(l)
```

```python
#The key and value representation
#It's ordered
d={
    "Virat":18000,
    "Rohit":15000,
    "Suresh":12000,
}
print(d)
```

```python
#Changable
d={
    "Virat":18000,
    "Rohit":15000,
    "Suresh":12000,
}
d['Suresh']=13000
print(d)
```

```python
#Duplicate will not be allowed
d={
    "Virat":18000,
    "Rohit":15000,
    "Suresh":12000,
    "Virat":20000,
}
print(d)
```

```python
#Insert in dictionary
d={
    "Virat":18000,
    "Rohit":15000,
    "Suresh":12000,
    "Virat":20000,
}
d["Rishav"]=14000
print(d)
```

```python
d={
    "Virat":18000,
    "Rohit":15000,
    "Suresh":12000,
    "Virat":20000,
}
d.update({
    "Rishav":14000,

})
print(d)
```

```python
d={
    "Virat":18000,
    "Rohit":15000,
    "Suresh":12000,
    "Rishav":20000,
}
for key in d.keys():
    print(key)
```

```python
import numpy as np
```

```python
#Creating the 1-D array
one=np.array([1,2,3,4,5,6])
print(one)
print(one.shape)
print(one.ndim)
```

```python
#Creating the 2-D array
two=np.array(
    [#Useless bracket
#    c1 c2 c3
     [1,2,3,10],#Row-1
     [4,5,6,11],#Row-2
     [7,8,9,12]#Row-3

    ]
)
print(two)
print(two.shape)
print(two.ndim)
```

```python
#Creating the 3-D array
three=np.array(
    [#Useless bracket
     [#Depth
      [1,2,3],
      [4,5,6],
      [7,8,9]

     ]

    ]

)
print(three)
print(three.ndim)
print(three.shape)
```

```python
#Slicing concept
#             -5 -4 -3 -2 -1
one=np.array([10,20,30,40,50])
#             0 ,1 ,2 ,3 , 4
'''
one[s:e:step]
'''
print(one[0:4:1])
print(one[-5:-2:1])
```

```python
two=np.array(
    [#Useless bracket
#    c1 c2 c3
     [1,2,3,10],#Row-0
#     0 1 2 3
     [4,5,6,11],#Row-1
#     0 1 2 3
     [7,8,9,12]#Row-2
#     0,1,2,3

    ]
)
'''
2,3,10
5,6,11

'''
'''
two[row_slicing,column_slicing]
'''
two[0:2,1:4]
```

```python
#Reshaping the array
one=np.array([1,2,3,4,5,6])
#3*2=3rows and 2 columns
#2*3=2rows and 3 columns
print(one.reshape(3,2))
print(one.reshape(2,3))
```

```python
#& operator case
one=np.array([10,20,30,40,50,60,70,80])
print(one[one<50])
print(one[one>50])
print(one[(one>30) & (one <70)])
```

```python
#Some inbuilt mathematical function in numpy
one=np.array([10,20,30,40,50,60,70,80])
print(one.min())
print(one.max())
print(one.sum())
print(one.mean())
print(np.median(one))
print(one.std())
print(one.var())
print(np.sqrt(one))
print(np.exp(one))
print(np.sin(one))
print(np.cos(one))
print(np.log(one))
print(np.tan(one))
```
